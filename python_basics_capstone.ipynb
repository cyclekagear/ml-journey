```python
marks = [45, 78, 88, 32, 90, 67, 55, 41, 99, 60]
print(max(marks))
print(min(marks))
print(len(marks))
sum = 0
for n in marks:
    sum+=n
print(sum)

    
```

```python
passed = 0
failed = 0
for n in marks:
    if n>=50:
        passed+=1
    else:
        failed+=1
print(passed, failed)
```

```python
for n in marks:
    if n>=85:
        print(n, 'A')
    elif 70<=n<=84:
        print(n, 'B')
    elif 50<=n<=69:
        print(n, 'C')
    else:
        print(n,'F')
```
